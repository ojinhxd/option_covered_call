```python
import pandas as pd
from datetime import date, timedelta

from pull_and_plot_covered_call_price import get_call_and_stock_close_price, plot_covered_call_price
```

```python
tickers = ['META','AMD','MSFT','PLTR','GLD','INTC','AMZN']
option_expiration_date = 'May 3, 2024'
# expiration_dates = op.get_expiration_dates(ticker)
```

```python
today = date.today()
stock_start_date = today - timedelta(5)  ## only needs to be several days ago so stock price pull does not return an empty dataset
```

```python
for ticker in tickers:
    try:
        data = get_call_and_stock_close_price(ticker,option_expiration_date,stock_start_date,today)
    except:
        print('=' * 100)
        print(f'{ticker} may not be valid. Please verify and try again!')
        continue
    data = get_call_and_stock_close_price(ticker,option_expiration_date,stock_start_date,today)    
    print('=' * 100)
    print(ticker)
    display(data)
    plot_covered_call_price(data,ticker)
    
print('=' * 100)
print('Covered call price data pull and plot completed successfully. Happy trading! :)')
```
